```python
import torch
from torch import nn
import torchvision
from PIL import Image

from gatys import Gatys
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
input_image = Image.open("content/Tuebingen_Neckarfront.jpg")
style_image = Image.open("style/vangogh_starry_night.jpg")
```

```python
gatys_method = Gatys().to(device)
```

```python
input_image.resize((200, 200))
```

```python
style_image.resize((200, 200))
```

```python
# random initialization
styled_input = gatys_method.transfer_style(input_image, style_image, opt_img=None)
```

```python
gatys_method.display_image(styled_input)
```

```python
from torch.autograd import Variable
# content initialization
opt_img = gatys_method.prep(input_image).unsqueeze(0).data.clone()
styled_input2 = gatys_method.transfer_style(input_image, style_image, opt_img=opt_img)
```

```python
gatys_method.display_image(styled_input2)
```

```python
# style initialization
styled_input3 = gatys_method.transfer_style(input_image, style_image, opt_img="style")
```

```python
gatys_method.display_image(styled_input3)
```
